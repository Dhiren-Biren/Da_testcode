# Combine two Transaction files

```python
import pandas as pd

import glob

path = r'C:\Users\STPL-SL04\Desktop\Code@test\test' 

all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:

    df = pd.read_csv(filename, index_col=None, header=0)

    li.append(df)

final_data = pd.concat(li, axis=0, ignore_index=False)
```

```python
final_data
```

```python
final_data.to_csv("FinalTr_Data.csv")
```
